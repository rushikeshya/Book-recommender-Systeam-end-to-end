## Popularity Based Recommender System

Our recommended system we can build on following Basis:
We recommended The Top 50 Books with highest average Rating but we will consider only those books which have got a minimum of 250 votes.

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.head()
```

```python
ratings_with_name.shape
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=250]
```

```python
popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

## Collaborative Filtering Based Recommender System

```python
ratings_with_name.head()
```

# Filtering on the basis of users

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x[x].index

expr_users = x[x].index
expr_users
```

The above Values show that every single user has made more than 200 votes, so we can say that they user's are most book reders, there are 811 users which would rating on above 200 books.

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(expr_users)]
```

```python
filtered_rating.shape
```

811 users made 474007 votes on books



# Filtering on the basis of Books

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

# Creating Recommender for Top 5 Books

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index== book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```

```python
recommend('1984')
```

```python
recommend("Animal Farm")
```

```python
recommend("The Hobbit : The Enchanting Prelude to The Lord of the Rings")
```
